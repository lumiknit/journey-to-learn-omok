### Agent & Policy Transformer

```python
def fn_name(fn):
  if hasattr(fn, 'name'): return fn.name
  else: return repr(fn)
  
def agent(pi, epsilon=0):
  # pi must return array of non-negative values
  def c(game):
    w, h = game.width, game.height
    m, p = np.ones(h * w), np.array(pi(game))
    for i in range(h * w):
      if game.board[i] != 0: m[i], p[i] = 0, 0
    s = p.sum()
    if np.random.uniform() < epsilon or s == 0:
      s = m.sum()
      if s == 0: return None, None # Cannot do anything
      else: idx = np.random.choice(h * w, p=(m / s))
    else: idx = np.random.choice(h * w, p=(p / s))
    return idx // w, idx % w
  c.name = 'stochastic({})'.format(fn_name(pi))
  return c

def softmax(arr, tau=1.0):
  arr = np.array(arr, dtype=np.float64)
  arr /= tau
  m = max(arr)
  z = np.exp(arr - m)
  return z / z.sum()

def pt_softmax(policy, tau=1.0):
  def p(game):
    p = policy(game)
    return softmax(p, tau=tau)
  p.name = 'pt_softmax({},tau={})'.format(fn_name(policy), tau)
  return p

def pt_norm(policy):
  def p(game):
    p = policy(game)
    return p / p.max()
  p.name = 'pt_norm({})'.format(fn_name(policy))
  return p

def policy_uniform(game):
  return np.ones(game.height * game.width)
policy_uniform.name = 'uniform'

def agent_mixed(game):
  a = np.random.uniform()
  if a < 0.4: return m5aa.agent(game)
  elif a < 0.8: return m5ad.agent(game)
  elif a < 0.9: return m5pt.agent(game)
  else: return m5df.agent(game)
agent_mixed.name = 'agent-mixed-analysis-based'
```

```python
def test_agents(num_game, agent1, agent2):
  w1, w2 = 0, 0
  c1, c2 = 0, 0
  for i in range(num_game):
    g = Mock5(H, W)
    result = g.play(agent1, agent2,
                    print_intermediate_state=False, print_messages=False)
    if result == 1:
      w1 += 1
      c1 += len(g.history)
    elif result == 2:
      w2 += 1
      c2 += len(g.history)
  print("-- Test Result --")
  print("* Agent1 = {} \n* Agent2 = {}".format(
      fn_name(agent1), fn_name(agent2)))
  print("Total : {:5d}".format(num_game))
  print("A1 Win: {:5d} ({:5.3f}) (avg.mov {:6.1f})".format(w1, w1 / num_game, c1 / max(1, w1)))
  print("A2 Win: {:5d} ({:5.3f}) (avg.mov {:6.1f})".format(w2, w2 / num_game, c2 / max(1, w2)))
```

### Neural Network

지금까지 policy network를 조금씩 수정해보면서 학습을 했는데,
현재까지의 결론은 다음과 같습니다.

- Kernel의 크기는 층마다 3 정도면 충분합니다. 조금 더 주변
정보를 반영하고자 5나 7로 늘려보았지만, 어차피 겹겹이 쌓이면
커버하는 범위는 비슷하게 됩니다.
- 층의 갯수는 신경망의 실행 속도와 비선형성에 직접 영향을 주므로
중요한 요소입니다.
오목의 경우에는 7개 정도만 되어도 적어도 알고리즘 상대로는
나쁘지 않은 수준의 학습이 가능해집니다.
AlphaGo가 CNN을 13층 쌓은 것처럼 조금 더 많이 쌓는 것도
좋겠지만, 학습시간과 타협을 해야하는 경우에는 조절이 필요합니다.
여기서는 CNN을 8층으로 쌓았습니다.
- Residual block을 사용하였습니다. 원래 residual block은
이미지를 다룰 때 사용한다고 알고 있는데, 오목판 역시 이미지인
만큼 사용해보면 어떨까 했습니다.
다만 실제로 그 차이는 그다지 체감되지 않는 듯 합니다.
- 다른 요소들보다 제일 중요한 것은 feature dimension입니다.
직관적으로는 돌 색 2가지, 연결될 수 있는 방향 4개,
돌 연결 개수 4개로 대강 $2 \times 4 \times 4 = 32$개의
feature만 있어도 될 것 같았지만, 실제로는 128개 밑으로
낮추면 제대로 학습이 되지 않습니다. 이 개수도
중간에 줄이게 되면 성능이 나빠지므로, 128개를 유지하는 정도는
해야됩니다.
실제로 제대로 학습하려면 256개 정도로 전부 채워야 하지 않을까
싶습니다.
- 아무래도 ReLU가 GELU에 비해서 계산량이 적지 않을까 해서
한번씩 시험을 해보는데, 여기서 학습을 할 때는 objective
function에서 미친듯이 널뛰기를 해서 local optimum을
넘어다녀야하는데 이 와중에 ReLU가 전부 죽어버리는 일이
종종 생깁니다. 그래서 결국에는 Leaky나 GELU 등을 사용하게
되는 것 같습니다.
- 기존에는 빈칸, 내 돌, 상대 돌로 총 3-dimension one hot
encoding을 사용했는데, 빈칸이 의미가 없지 않나 싶어서
현재는 입력에서 제외했습니다. 즉 내 돌, 상대 돌로 2-dimension
one hot encoding을 입력으로 받습니다.

```python
class Flatten(nn.Module):
  def forward(self, x):
    if len(x.shape) == 3: return x.view(-1)
    else: return x.flatten(1, -1)

class Block1(nn.Module):
  def __init__(self, ch, int_ch, ker):
    super().__init__()
    self.seq = nn.Sequential(
      nn.Conv2d(ch, int_ch, ker, padding='same'),
      nn.GELU(),
      nn.Conv2d(int_ch, ch, ker, padding='same'))
  def forward(self, x):
    y_0 = self.seq(x)
    return nn.functional.gelu(x + y_0)

class Policy(nn.Module):
  def __init__(self):
    super().__init__()
    self.seq = nn.Sequential(
      nn.Conv2d(2, 128, 3, padding='same'),
      nn.GELU(),
      Block1(128, 128, 3),
      Block1(128, 128, 3),
      Block1(128, 128, 3),
      nn.Conv2d(128, 1, 5, padding='same'),
      nn.GELU(),
      # Flatten
      Flatten(),
      # Softmax
      nn.LogSoftmax(dim=-1))

  def forward(self, x):
    return self.seq(x)

def init_nn():
  global policy, policy_backs
  policy = Policy().to(device)
  policy_backs = []
  save_policy()

def save_policy():
  global policy_backs
  n = Policy().to(device)
  n.load_state_dict(policy.state_dict())
  policy_backs = (policy_backs + [n])[-10: ]
```

```python
def game_to_tensor(game):
  t = torch.zeros(2, game.height * game.width, dtype=torch.float)
  for i in range(game.height * game.width):
    b = game.board[i]
    if b == game.player: t[0][i] = 1.0
    elif b != 0: t[1][i] = 1.0
  return t.view(2, game.height, game.width)
```

```python
def policy_model(net):
  def c(game):
    X = game_to_tensor(game).to(device)
    with torch.no_grad():
      p = torch.exp(net(X))
    arr = p.squeeze().to('cpu').numpy()
    return arr
  c.name = 'model({:x})'.format(id(net))
  return c
```

```python
class Stat:
  def __init__(self, n=0, w=0, l=0, d=0): self.n, self.w, self.l, self.d = n, w, l, d
  def dup(self):
    return Stat(self.n, self.w, self.l, self.d)
  def __sub__(self, other):
    return Stat(self.n - other.n, self.w - other.w, self.l - other.l, self.d - other.d)
  def win(self):
    self.n += 1
    self.w += 1
  def lose(self):
    self.n += 1
    self.l += 1
  def draw(self):
    self.n += 1
    self.d += 1
  def update_by_result(self, result):
    self.n += 1
    if result == 1: self.w += 1
    elif result > 1: self.l += 1
    else: self.d += 1
  def win_rate(self):
    if self.n >= 1: return (self.w + 0.5 * self.d) / self.n
    else: return np.nan
```

```python
def append_game_replay(Xs, As, Vs, gamma, game, result, winner_only=True):
  reward = 1 if result != 0 else 0.3
  # Append to Batch
  for f in range(2):
    for r in range(4):
      g = game.replay(angle=r, flip=f)
      while len(g.history) > 0:
        a = g.history[-1]
        g.undo()
        if not winner_only or g.player == 3 - game.player:
          Xs.append(game_to_tensor(g))
          As.append(a)
          Vs.append(reward)
        reward *= gamma
```

```python
import os
from read_record import file_to_int_pair_iter

def gen_episode_by_generator(num_game, rnd):
  def generator(Xs, As, Vs, gamma, stat):
    filename = "temp.data"
    os.system("./gen {} {} {} {} {}".format(filename, num_game, H, W, rnd))
    i_iter = file_to_int_pair_iter(filename)
    left = 0
    game, result = None, None
    for i0, i1 in i_iter:
      if left <= 0:
        game = Mock5(H, W)
        result, left = i0, i1
      else:
        left -= 1
        game.place_stone_at_index(i0)
      if left == 0:
        if result != game.check_win():
          raise ValueError
        append_game_replay(Xs, As, Vs, gamma, game, result, winner_only=False)
    return num_game
  return generator
```

```python
def gen_episode_by_play(
    agent1,
    agent2
):
  def generator(Xs, As, Vs, gamma, stat):
    # Run Game
    game = Mock5(H, W)
    result = game.play(agent1, agent2,
      print_intermediate_state=False, print_messages=False)
    # Make reward
    stat.update_by_result(result)
    append_game_replay(Xs, As, Vs, gamma, game, result)
    return 1
  return generator
```

```python
# REINFORCE
def learn(
    opt,
    n_episode,
    gen_episode, #: take (Xs, As, Vs, gamma, stat)
    n_epoch,
    gamma,
    batch_size,
    interval_stat
):
  stat = Stat()
  last_stat = stat.dup()
  last_stat_epi = 0
  loss_inc = 0
  Xs, As, Vs = [], [], []
  epi = 0
  while epi < n_episode:
    # Generate episode
    inc_epi = gen_episode(Xs, As, Vs, gamma, stat)
    if type(inc_epi) is int: epi += inc_epi
    # If batch is full enough, perform gradient ascent
    if len(Xs) >= batch_size:
      # Tensor-fy
      X = torch.stack(Xs)
      X += (torch.rand(X.shape) - 0.5) * 0.01
      X = X.to(device)
      A = torch.tensor(As).unsqueeze(dim=1).to(device)
      V = torch.tensor(Vs, dtype=torch.float).to(device)
      Xs, As, Vs = [], [], []
      # Learn
      loss_list = []
      for e in range(n_epoch):
        opt.zero_grad(set_to_none=True)
        pi_s = policy(X)
        pi_sa = pi_s.gather(1, A).squeeze(dim=1)
        loss = -(V * pi_sa).mean()
        loss.backward()
        opt.step()
        loss_list.append(loss.mean().item())
      loss_inc += loss_list[0] - loss_list[-1]
      # Print status and evaluate
      if epi - last_stat_epi >= interval_stat:
        save_policy()
        print("----------")
        print("Ep #{:<6d} Loss Change Accum {:13.10f}".format(
          epi, loss_inc))
        print("  Win Rate {:8.4f}% ({}w + {}d + {}l = {})".format(
            100 * stat.win_rate(), stat.w, stat.d, stat.l, stat.n))
        dstat = stat - last_stat
        if dstat.n > 0:
          print("   WR Diff {:8.4f}% ({}w + {}d + {}l = {})".format(
              100 * dstat.win_rate(), dstat.w, dstat.d, dstat.l, dstat.n))
        else:
          test_a = agent(policy_model(policy))
          #test_agents(10, test_a, m5rand.agent)
          test_agents(10, test_a, m5pt.agent)
        last_stat, last_stat_epi = stat.dup(), epi
        loss_inc = 0
```

### Learning

- 처음에는 기보만 1500개 넣고 학습을 시킨 뒤, 이후에는
`agent_mixed`와 대국하면서 배웁니다.
- 학습은 10시간 정도 지난 뒤에 강제로 interrupt하였습니다.
- Adam은 중간에 loss가 아예 날아가거나 특정 지점에서 멈춰버리는
일이 있어 Momentum SGD만을 사용합니다.

```python
def run():
  init_nn()

  timer = time.process_time()

  opt = optim.SGD(policy.parameters(),
                  lr=1e-2,
                  momentum=0.8,
                  weight_decay=1e-6)
  
  learn(
      opt = opt,
      n_episode = 1500,
      gen_episode = gen_episode_by_generator(1, np.random.uniform() * 0.04),
      n_epoch = 2,
      gamma = 0.995,
      batch_size = 1,
      interval_stat = 250)
  
  print("\n#-- Pre Train : {:.4f} sec elapsed\n".format(time.process_time() - timer))
  timer = time.process_time()
  
  opt = optim.SGD(policy.parameters(),
                  lr=1e-2,
                  momentum=0.8,
                  weight_decay=1e-6)

  agent1 = agent(policy_model(policy))
  agent2 = agent_mixed
  learn(
      opt = opt,
      n_episode = 1350000,
      gen_episode = gen_episode_by_play(agent1, agent2),
      n_epoch = 2,
      gamma = 0.995,
      batch_size = 50,
      interval_stat = 500)
  
  print("\n#-- Main Train : {:.4f} sec elapsed\n".format(time.process_time() - timer))
  timer = time.process_time()

run()
```
